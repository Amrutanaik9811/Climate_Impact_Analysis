
## Key-Crop-Yields Dataset

## Description

The **Key-Crop-Yields** dataset contains information on crop yields across various countries and years. The dataset includes multiple crop yield values, typically for staple crops like wheat, rice, maize, and others. This data helps in analyzing global food production trends, climate impacts, and agricultural productivity.

## Data Summary

- The dataset records agricultural productivity over time, primarily focusing on key staple crops.
- It provides historical crop yield data from 1961 onwards, giving insights into how crop yields have changed over the decades in various regions.

```python
# Read the CSV file
yeilds = pd.read_csv("key-crop-yields.csv")

# Filter data to include only continents and World
entities_to_keep = ["World", "Asia", "Europe", "North America", "South America", "Africa", "Oceania", "Antarctica"]
yeilds = yeilds[yeilds['Entity'].isin(entities_to_keep)]

# Drop the 'Code' column if it exists
if 'Code' in yeilds.columns:
    yeilds = yeilds.drop(columns=['Code'])

# Clean 'Year' column: Ensure it's numeric and drop invalid data
if 'Year' in yeilds.columns:
    yeilds['Year'] = pd.to_numeric(yeilds['Year'], errors='coerce')  # Convert non-numeric to NaN
    yeilds = yeilds.dropna(subset=['Year'])  # Drop rows with missing 'Year'
    yeilds['Year'] = yeilds['Year'].astype(int)  # Convert 'Year' to integers
else:
    raise ValueError("The 'Year' column is missing in the dataset.")

yeilds = yeilds.fillna(0)

# Exclude all-zero columns to keep meaningful data
yeilds = yeilds.loc[:, (yeilds != 0).any(axis=0)]  # Keep columns with at least one non-zero value

# Normalize all numeric columns except 'Year' using MinMaxScaler
scaler = MinMaxScaler()


numeric_columns = yeilds.select_dtypes(include=['number']).columns
numeric_columns = [col for col in numeric_columns if col != 'Year']  # Exclude 'Year'

# Apply normalization
yeilds[numeric_columns] = scaler.fit_transform(yeilds[numeric_columns])

# Reset the index to start from 0
yeilds = yeilds.reset_index(drop=True)

# Display the cleaned and normalized data

display(HTML(yeilds.to_html(border=1)))  
output_path = "keycropyeildsnormal"
```

## Change in Temperature caused with greenhouse gas emissions by land and agriculture Dataset

## Description
This dataset provides information on temperature rise associated with greenhouse gas emissions by land and agriculture. It includes data related to how temperature anomalies are influenced by various factors like greenhouse gas emissions, with a focus on global warming trends over the years.

## Data Summary
This dataset is important for understanding the impact of land use and agriculture on global temperature changes, providing insights into how different regions contribute to the warming of the planet.

```python
dest = "temperature-rise-with-country-codes.csv"

# Load the dataset
temp_change = pd.read_csv(dest)

# Filter for records starting from year 1961
temp_change = temp_change[temp_change["Year"] >= 1961]

# Drop the 'Code' column if it exists
if 'Code' in temp_change.columns:
    temp_change = temp_change.drop(columns=['Code'])

# Filter to include only continents and World
entities_to_keep = ["World", "Asia", "Europe", "North America", "South America", "Africa", "Oceania", "Antarctica"]
temp_change = temp_change[temp_change['Entity'].isin(entities_to_keep)]

# Replace NaN values with 0
temp_change = temp_change.fillna(0)

# Reset the index after filtering
temp_change.reset_index(drop=True, inplace=True)


numeric_cols = temp_change.select_dtypes(include=['float64', 'int64']).columns.difference(['Year'])


if not numeric_cols.empty:
    scaler = MinMaxScaler()
    temp_change[numeric_cols] = scaler.fit_transform(temp_change[numeric_cols])
    print("\nData has been normalized (Min-Max scaling applied).")
else:
    print("\nNo numeric columns found to normalize.")

# Display the first 10 rows with styling and table borders
x = temp_change.head(10)
print("Filtered and Normalized Temperature Rise Data (Continents + World):")
display(HTML(x.to_html(border=1)))


output_path = "temprisenormal.csv"
temp_change.to_csv(output_path, index=False, float_format='%.6f')

print(f"\nCleaned and normalized data saved to: {output_path}")
```

This dataset tracks the global precipitation anomaly, showing how precipitation levels deviate from the average over time. Precipitation anomalies are essential for understanding shifts in weather patterns, including droughts and floods, which can be influenced by various climatic factors.

```python
file_path = "global-precipitation-anomaly.csv"

percipitation = pd.read_csv(file_path)

# Drop the 'Code' column if it exists
percipitation = percipitation.drop(columns=['Code'], errors='ignore')

# Filter for records starting from year 1961
percipitation = percipitation[percipitation["Year"] >= 1961]

# Replace NaN values with 0
percipitation = percipitation.fillna(0)

# Reset the index after filtering
percipitation.reset_index(drop=True, inplace=True)

# Identify numeric columns (excluding 'Year' if present)
column = percipitation.select_dtypes(include=['float64', 'int64']).columns.difference(['Year'])

# Apply Min-Max Normalization
if not column.empty:
    scaler = MinMaxScaler()
    percipitation[column] = scaler.fit_transform(percipitation[column])
    print("\nData has been normalized (Min-Max scaling applied).")
else:
    print("\nNo numeric columns found to normalize.")


percipitation.head(10)






```

## This below code is designed to read multiple CSV files, merge them into a single dataset, clean and preprocess the data, normalize numerical columns, and finally save the processed dataset into a new CSV file. Heres a breakdown of the steps and operations performed:

**Merging Logic:** :This part starts by defining the base dataset as "key_crop_yields". The merge_order list contains the order in which the datasets will be merged, specifying the key columns for each dataset ("Entity", "Code", and "Year"). The code then iteratively merges each dataset into the merged_data DataFrame using the pd.merge() function with an outer join to keep all records from both datasets. To avoid column name conflicts, a suffix is added to the column names from the merged datasets.

**Removing Duplicates:** After merging the data, duplicate rows (if any) are removed using drop_duplicates().

**Handling Missing Values:** Missing values are handled based on the data type:
    --For categorical columns (dtype object), missing values are replaced with the string "Unknown".
    --For numerical columns, missing values are replaced with 0.

**Normalization:** This step scales the numerical data to a range between 0 and 1 using MinMaxScaler from scikit-learn. It first identifies the numeric columns (excluding the "Year" column), then applies the scaler to normalize these columns. This helps standardize the data, which is essential for comparison and further analysis.

**Ensuring Float Format:** After normalization, the columns are explicitly converted to numeric types (floats) to avoid any inconsistencies in the data.



```python
files = {
    "key_crop_yields": "key-crop-yields.csv",
    "temperature_rise_with_country_codes": "temperature-rise-with-country-codes.csv",
    "annual_temperature_anomalies": "annual-temperature-anomalies.csv",
    "global_precipitation_anomaly": "global-precipitation-anomaly.csv",
}

# List of major continents + "World"
entities_to_keep = ["World", "Asia", "Europe", "North America", "South America", "Africa", "Oceania", "Antarctica"]

# Load the datasets, filter by year >= 1961, and remove 'Code' column
data = {}
for name, path in files.items():
    try:
        df = pd.read_csv(path)
        
        # Filter to include only years starting from 1961
        if "Year" in df.columns:
            df = df[df["Year"] >= 1961]
        
        # Drop 'Code' column if it exists
        if "Code" in df.columns:
            df = df.drop(columns=["Code"])
        
        data[name] = df
        print(f"{name} daata has been loaded and removed the values from1961.")
    except Exception as e:
        print(f"Error loading {name} from {path}: {e}")

# Start merging datasets using 'key_crop_yields' as the base
datamerging = data["key_crop_yields"]

# Define merge order
merge_order = [
    ("temperature_rise_with_country_codes", ["Entity", "Year"]),
    ("annual_temperature_anomalies", ["Entity", "Year"]),
    ("global_precipitation_anomaly", ["Entity", "Year"]),
]

# Merge with other datasets step-by-step
for dataset_name, keys in merge_order:
    if dataset_name in data:
        try:
            datamerging = pd.merge(
                datamerging,
                data[dataset_name],
                on=keys,
                how="outer",
                suffixes=('', f'_{dataset_name}')  
            )
            print(f"Merged with {dataset_name} successfully.")
        except KeyError as e:
            print(f"KeyError during merging {dataset_name}: {e}")
        except Exception as e:
            print(f"Error merging {dataset_name}: {e}")


datamerging = datamerging.drop_duplicates()


for column in datamerging.columns:
    if datamerging[column].dtype == "object":
        datamerging[column].fillna("Unknown", inplace=True)
    else:
        datamerging[column].fillna(0, inplace=True)


column = datamerging.select_dtypes(include=['float64', 'int64']).columns

# Exclude 'Year' column from the normalization
column = [col for col in column if col != "Year"]


scaler = MinMaxScaler()


datamerging[column] = scaler.fit_transform(datamerging[column])

# Ensure numeric columns remain as floats after normalization
for col in column:
    datamerging[col] = pd.to_numeric(datamerging[col], errors='coerce')

# Filter data to keep only the major continents and 'World'
filtered_data = datamerging[datamerging['Entity'].isin(entities_to_keep)]

# Save the filtered data to a CSV file
output_path = "finalpp.csv"
filtered_data.to_csv(output_path, index=False, float_format='%.6f')  # Float format ensures proper precision



filtered_data.describe()
```

```python
data = pd.read_csv('finalpp.csv')

# 1. Create a total crop production feature
crop_columns = ['Wheat ', 'Rice', 'Bananas ', 'Maize ', 'Soybeans ', 'Potatoes', 
                'Beans, dry ', 'Peas, dry ', 'Cassava ', 'Cocoa beans', 'Barley ']
data['Total_Crop_Production'] = data[crop_columns].sum(axis=1)
print("adding Total_Crop_Production:")
data[['Entity', 'Year', 'Total_Crop_Production']].head()


```

```python
# 2. Calculate year-over-year change in temperature anomaly
data['Temperature_Change'] = data.groupby('Entity')['Temperature anomaly'].diff()
print("After adding Temperature_Change:")
data[['Entity', 'Year', 'Temperature anomaly', 'Temperature_Change']].head()
```

```python
# 3. Create a feature for cumulative temperature anomaly
data['Cumulative_Temperature_Anomaly'] = data.groupby('Entity')['Temperature anomaly'].cumsum()
print("After adding Cumulative_Temperature_Anomaly:")
data[['Entity', 'Year', 'Temperature anomaly', 'Cumulative_Temperature_Anomaly']].head()

```

```python
# 4. Calculate the ratio of temperature anomaly to precipitation anomaly
data['Temp_Precip_Ratio'] = data['Temperature anomaly'] / data['Annual precipitation anomaly']
print("After adding Temp_Precip_Ratio:")
data[['Entity', 'Year', 'Temperature anomaly', 'Annual precipitation anomaly', 'Temp_Precip_Ratio']].head()

```

```python
# 5. Create binary features for extreme weather conditions
data['Extreme_Heat'] = (data['Temperature anomaly'] > data['Temperature anomaly'].quantile(0.95)).astype(int)
data['Extreme_Precipitation'] = (data['Annual precipitation anomaly'] > data['Annual precipitation anomaly'].quantile(0.95)).astype(int)
print("After adding Extreme_Heat and Extreme_Precipitation:")
data[['Entity', 'Year', 'Temperature anomaly', 'Annual precipitation anomaly', 'Extreme_Heat', 'Extreme_Precipitation']].head()
```

```python
# 6. Calculate rolling averages for temperature and precipitation anomalies
data['Temp_Anomaly_5yr_Avg'] = data.groupby('Entity')['Temperature anomaly'].rolling(window=5).mean().reset_index(0, drop=True)
data['Precip_Anomaly_5yr_Avg'] = data.groupby('Entity')['Annual precipitation anomaly'].rolling(window=5).mean().reset_index(0, drop=True)
print("After adding 5-year rolling averages:")
data[['Entity', 'Year', 'Temperature anomaly', 'Annual precipitation anomaly', 'Temp_Anomaly_5yr_Avg', 'Precip_Anomaly_5yr_Avg']].head()

```

```python
# 7. Create interaction terms between temperature anomaly and crop production
for crop in crop_columns:
    data[f'{crop}_Temp_Interaction'] = data[crop] * data['Temperature anomaly']
print("After adding crop-temperature interactions:")
data[['Entity', 'Year'] + [f'{crop}_Temp_Interaction' for crop in crop_columns]].head()


```

```python
# 8. Calculate the percentage change in crop production
for crop in crop_columns:
    data[f'{crop}_Pct_Change'] = data.groupby('Entity')[crop].pct_change()

print("After adding percentage change in crop production:")
data[['Entity', 'Year'] + [f'{crop}_Pct_Change' for crop in crop_columns]].head()

```

```python
# 9. Create a feature for years since 1961 (the start of the dataset)
data['Years_Since_1961'] = data['Year'] - 1961
print("After adding Years_Since_1961:")
data[['Entity', 'Year', 'Years_Since_1961']].head()

```

```python
# 10. Encode the 'Entity' column using one-hot encoding
entity_encoded = pd.get_dummies(data['Entity'], prefix='Entity')
data = pd.concat([data, entity_encoded], axis=1)

print("After one-hot encoding Entity:")
data[['Entity'] + [col for col in data.columns if col.startswith('Entity_')]].head()


# Save the engineered dataset
data.to_csv('engineered_dataset.csv', index=False)

print("Feature engineering completed. Dataset saved as 'engineered_dataset.csv'.")
```

## Linear Regression Model :
# Interpreting the Results
Key Outputs

Coefficients (1,2):

Check the sign of the coefficients:

Negative () for temperature/precipitation  Reduced yields.
Positive (+) for temperature/precipitation  Beneficial yields.
p-values: Determine if coefficients are statistically significant (p<0.05).
R: Check how well the model fits the data.

Example Interpretation
Wheat:1=0.05, <0.01 : Significant negative impact of temperature on wheat yields.
2=  0.02, <0.05 : Significant negative impact of precipitation on wheat yields.
Rice:1=+0.03, <0.01: Significant positive impact of temperature on rice yields.
2=0.01, >0.05: No significant impact of precipitation on rice yields.

This below model is used to plot wheat maize and rice production over the years 

```python
df['Year'] = pd.to_datetime(df['Year'], format='%Y')
global_yields = df.groupby('Year')[['Wheat ', 'Maize ', 'Rice']].mean().reset_index()


X = global_yields['Year'].dt.year.values.reshape(-1, 1)
y_wheat = global_yields['Wheat '].values
y_maize = global_yields['Maize '].values
y_rice = global_yields['Rice'].values

crops = ['Wheat ', 'Maize ', 'Rice']
colors = ['r', 'g', 'b']
plt.figure(figsize=(12, 8))

for crop, y, color in zip(crops, [y_wheat, y_maize, y_rice], colors):
    model = LinearRegression()
    model.fit(X, y)
    y_pred = model.predict(X)
    
    # Calculate R-squared
    r2 = r2_score(y, y_pred)
    
    # Plot the data and regression line
    plt.scatter(X, y, color=color, alpha=0.5, label=f'{crop} (R = {r2:.2f})')
    plt.plot(X, y_pred, color=color, linewidth=2)
    
    # Print the slope (trend)
    print(f"{crop} yield trend: {model.coef_[0]:.4f} per year")

plt.xlabel('Year')
plt.ylabel('Yield')
plt.title('Global Crop Yields Over Time')
plt.legend()
plt.grid(True)
plt.show()
```

# This file is used to plot the tempereture anamoly and crop yeilds 

```python

df = pd.read_csv('finalpp.csv')

# Convert 'Year' to datetime
df['Year'] = pd.to_datetime(df['Year'], format='%Y')

# List of all entities
entities = df['Entity'].unique()

# List of crops
crops = ['Wheat ', 'Maize ', 'Rice']

# Create a figure with subplots for each entity
fig, axs = plt.subplots(len(entities), 1, figsize=(12, 6*len(entities)), sharex=True)
fig.suptitle('Crop Yields vs Annual Precipitation Anomaly by Entity')

for i, entity in enumerate(entities):
    entity_data = df[df['Entity'] == entity]
    
    # Prepare data for linear regression
    X = entity_data['Annual precipitation anomaly'].values.reshape(-1, 1)
    
    for crop, color in zip(crops, ['r', 'g', 'b']):
        y = entity_data[crop].values
        
        model = LinearRegression()
        model.fit(X, y)
        y_pred = model.predict(X)
        
        # Calculate R-squared
        r2 = r2_score(y, y_pred)
        
        # Plot the data and regression line
        axs[i].scatter(X, y, color=color, alpha=0.5, label=f'{crop} (R = {r2:.2f})')
        axs[i].plot(X, y_pred, color=color, linewidth=2)
        
        # Print the slope (trend)
        print(f"{entity} - {crop} yield trend: {model.coef_[0]:.4f} per unit change in precipitation anomaly")
    
    axs[i].set_ylabel('Yield')
    axs[i].set_title(entity)
    axs[i].legend()
    axs[i].grid(True)

plt.xlabel('Annual Precipitation Anomaly')
plt.tight_layout()
plt.show()
```
